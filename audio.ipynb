```python
import os
import numpy as np
import librosa
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import layers, models, callbacks
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
import librosa.effects as effects
import tensorflow as tf

# Function to extract features from audio files
def extract_features(file_path, mfcc=True, chroma=True, mel=True):
    audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    features = []
    
    if mfcc:
        mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40).T, axis=0)
        features.append(mfccs)
    if chroma:
        chroma = np.mean(librosa.feature.chroma_stft(y=audio, sr=sample_rate).T, axis=0)
        features.append(chroma)
    if mel:
        mel = np.mean(librosa.feature.melspectrogram(y=audio, sr=sample_rate).T, axis=0)
        features.append(mel)
    
    return np.concatenate(features)

# Example of data augmentation
def augment_data(audio, sample_rate):
    augmented_audio = effects.pitch_shift(audio, sample_rate, n_steps=2)  # Adjust the value as needed
    return augmented_audio

def lr_scheduler(epoch, lr):
    if epoch % 5 == 0:
        return lr * 0.9  # Adjust the factor as needed
    else:
        return lr

lr_callback = LearningRateScheduler(lr_scheduler)

# Function to load and preprocess the dataset
def load_data(data_dir, split='train',augment=True):
    labels = []
    features = []

    for folder in os.listdir(data_dir):
        subfolder = os.path.join(data_dir, folder)
        for file_name in os.listdir(subfolder):
            file_path = os.path.join(subfolder, file_name)

            audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
            if augment:
                audio = augment_data(audio, sample_rate)


            feature = extract_features(file_path)
            features.append(feature)
            labels.append(folder)

    X = np.array(features)
    y = np.array(labels)

    # Encode labels
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    if split == 'train':
        return X, y, label_encoder.classes_
    elif split == 'test':
        return X, y

# Load and preprocess the data
train_data_dir = r'C:/Users/MOMI/Desktop/project/train'
test_data_dir = r'C:/Users/MOMI/Desktop/project/test'

X_train, y_train, classes = load_data(train_data_dir, split='train',augment=True)
X_test, y_test = load_data(test_data_dir, split='test',augment=True)

# Reshape the input data for CNN
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Build the CNN model
model = models.Sequential()
model.add(layers.Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(len(classes), activation='softmax'))
model.add(layers.BatchNormalization())

# Build the CNN model with additional layers
model = models.Sequential()
model.add(layers.Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Conv1D(128, kernel_size=3, activation='relu'))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Conv1D(256, kernel_size=3, activation='relu'))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))  # Add dropout to reduce overfitting
model.add(layers.Dense(len(classes), activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model.fit(X_train, y_train, epochs=20, validation_data=(X_test, y_test), callbacks=[early_stop, lr_callback])

# Save the model
model.save('bird_voice_classification_model.h5')

# Evaluate the model
accuracy = model.evaluate(X_test, y_test)[1]
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import numpy as np
import librosa
from tensorflow.keras.models import load_model
import sounddevice as sd

# Load the trained model
model = load_model('bird_voice_classification_model.h5')

# Define your classes
classes = ['Asian Koel', 'Black Kite', 'Black-breasted Parrotbill', 'Black-necked crane',
           'Bristled Grassbird', 'Cheer Pheasant', 'Common Cuckoo', 'Common Pochard',
           'Common Wood Pigeon', 'Eastern Imperial Eagle', 'Egyptian Vulture',
           'Great Slaty Wookpecker', 'Greater Spotted Eagle', 'Grey Treepie',
           'Grey-crowned Prinia', 'Grey-sided Thrush', 'Himalayan Monal', 'House Crow',
           'House Sparrow', 'Indian Spotted Eagle', 'Jerdon_s Babbler', 'Kashmir Flycatcher',
           'Large-billed Crow', 'Long-tailed Duck', 'Pallas_s Fish Eagle', 'Red-billed Blue Magpie',
           'Rose-ringed Parakeet', 'Rufous Treepie', 'Rufous-necked Hornbill', 'Saker Falcon',
           'Sarus Crane', 'Satyr Tragopan', 'Slender-billed Babbler', 'Spiny Babbler',
           'Spotted Dove', 'Steppe Eagle', 'Swamp Francolin', 'Swamp Grass-Babbler',
           'White-throated Bushcat', 'Wood Snipe']

# Function to extract features from real-time audio
def extract_realtime_features(audio_chunk, sample_rate=44100):
    audio_chunk = audio_chunk.astype(np.float32) / 32767.0  # Convert integer to float and normalize
    mfccs = np.mean(librosa.feature.mfcc(y=audio_chunk, sr=sample_rate, n_mfcc=40).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=audio_chunk, sr=sample_rate).T, axis=0)
    mel = np.mean(librosa.feature.melspectrogram(y=audio_chunk, sr=sample_rate).T, axis=0)

    return np.concatenate([mfccs, chroma, mel]).astype(np.float32)


# Function to capture audio for a specific duration and make predictions
def capture_and_predict():
    duration = 5  # Set a shorter duration for real-time capture
    sample_rate = 44100
    chunk_size = 1024
    confidence_threshold = 0.5  # Adjust as needed

    print(f"Listening to a {duration}-second bird sound...")

    try:
        # Capture audio in chunks
        audio_stream = sd.InputStream(samplerate=sample_rate, channels=1, dtype=np.int16)
        audio_stream.start()

        audio_data = np.zeros(int(duration * sample_rate), dtype=np.int16)
        for i in range(0, int(sample_rate * duration), chunk_size):
            audio_chunk, overflowed = audio_stream.read(chunk_size)
            audio_chunk = audio_chunk.flatten()[:audio_data[i:i+chunk_size].shape[0]]
            audio_data[i:i+chunk_size] = audio_chunk

        audio_stream.stop()

        # Extract features from the captured audio
        features = extract_realtime_features(audio_data, sample_rate)

        # Reshape the input data for the CNN model
        input_data = features.reshape(1, features.shape[0], 1)

        # Make predictions
        predictions = model.predict(input_data)

        # Check confidence level
        if np.max(predictions) > confidence_threshold:
            predicted_class = np.argmax(predictions)
            print(f"Predicted Bird Species: {classes[predicted_class]}")
        else:
            print("Prediction not confident enough.")

    except KeyboardInterrupt:
        print("Stopped listening.")

# Run the audio capture and prediction
capture_and_predict()
```
